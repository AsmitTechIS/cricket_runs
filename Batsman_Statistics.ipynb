# Most Ducks

```python
mostducks=temp[['Player', '0']].sort_values(by="0",ascending=False)

plt.figure(figsize=(8,8))
graph=sns.barplot(y='Player',palette='rocket',x='0',data=mostducks.head(25))
graph.set_title('Most Ducks By a Batsman')
plt.show()
```

# Most Sixes

```python
mostsixes=temp[['Player', '6s']].sort_values(by="6s",ascending=False)
plt.figure(figsize=(8,8))
graph=sns.barplot(y='Player',palette='rocket',x='6s',data=mostsixes.head(25))
graph.set_title('Most Sixes By a Batsman')
plt.show()
```
